```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special
from scipy.stats import multivariate_normal
import pandas as pd

class babyHTM(object):

    """
    Implementation of anomaly detection using HTM algorithm described in "Unsupervised real-time anomaly detection 
    for streaming data" - Ahmad, Lavin, Purdy, Agha *2017
    As well as Numenta Spatial Pooling and Temporal Memory pseudocode documents

    Parameters
    ----------
    a        : sparse representation of input vectors (size[T,N])
    pi       : HTM prediction of input vectors (size[T,N])
    case     : matrix state between input and prediction ["exact","random","nearexact"]
    N        : number of HTM network columns - vector dimension
    T        : max time as matrix dimension - matrix dimension
    sparsity : percentage of on bits in array
    j            : input space unit (neuron)
    i            : SP mini column array unit
    gamma        : hyperspace dimension ("input space receptive field")
    N_pot_syn    : number of potential synapses on each dendritic segment
    N_input      : input space dimension
    N_columns    : columns space dimension
    connectivity : initial connectivity percentage
    p_thresh     : permanence threshold
    R            : input inhibition radius

    * - refers to differences between code description and function
    
    """

    def __init__(self, 
                 N_input = 30,connectivity = .8,p_thresh = 0.7,
                 N=2048, T=50, case="random", sparsity=2.,R=5,cellsPerColumns=16,
                 N_column = 30,PERMANENCE_DECREMENT=0.2, PERMANENCE_INCREMENT=0.2):
        """
        Variable storage
        """
        self.N_input = N_input         
        self.connectivity = connectivity 
        self.p_thresh = p_thresh
        self.N = N
        self.T = T
        self.case = case
        self.sparsity = sparsity
        self.pi = []
        self.a = []
        self.R = R 
        self.cellsPerColumns = cellsPerColumns
        self.N_column = N_column 
        self.HTMlayers = np.zeros((self.N_column ,self.cellsPerColumns))
        self.synapses = pd.DataFrame({'permanences':[], 'presynapticCell':[]})
        self.segment = pd.DataFrame({'synapses':[], 'cell':[]})
        self.activeCells = []
        self.winnerCells = []
        self.matchingSegments = [1,1,1,1,1]
        self.PERMANENCE_INCREMENT = PERMANENCE_INCREMENT
        self.PERMANENCE_DECREMENT = PERMANENCE_DECREMENT
        self.potentialSynapses = np.zeros((N_column,N_column,R*2,R*2))
        self.columns = np.zeros((N_column,N_column))
 
    def run_model(self):
        """
        Function which calls other functions and structures overall model routine
        """
        # ----- ANALYSIS METHODS -----
        #self.random_input()
        #self.prediction_error()
        #self.anomaly_likelihood()
        # ----- HTM STEPS ----- eventually to run over time
        self.SP_phase1_initizlization(printer="Y") # printer="Y" - prints outputs
        self.SP_phase2_overlap()
        self.SP_phase3_competition()
        self.SP_phase4_update()
        #for t in np.arange(1,5,1):
            #self.TM_phase1_context(t,printer="Y")
            #self.TM_phase2_prediction()

    def random_input(self):
        """
        Randomly generate binary arrays with sparse distribution
        """
        a = np.random.choice([1, 0], size=(self.T,self.N), p=[self.sparsity/100, (100-self.sparsity)/100])

        # Specify simulated prediction matrix
        if self.case == "exact":
            pi = a
        elif self.case == "random":
            pi = np.random.choice([1, 0], size=(self.T,self.N), p=[self.sparsity/100, (100-self.sparsity)/100])
        elif self.case == "nearexact": #########TBC
            pi = a
            pi[1]=(pi[1]*-1)+1
        else:
            print("Case key error")

        self.pi = pi
        self.a = a

    def prediction_error(self, printer="n"):
        """
        calculate prediction error
        """
        S = np.zeros(self.T)

        # Loop across rows
        for t in np.arange(self.T):
            # Commonality between matricies
            top = np.dot(self.pi[t,:],self.a[t,:]) 

            # Total number of bits in ISM
            bottom = np.sum(self.a[t,:]) # not norm*

            # Calculate Prediction error
            S[t] = 1 - (top/(bottom))

            ## Check rows
            #if S[t] == 0:
            #    print("Row =", t+1, "Perfect match : S = 0")
            #else:
            #    print("Row =", t+1, "Inexact match with : S =", S[t])
        self.S = S
        
        if printer == "Y":
            plt.plot(S)
            plt.xlabel('Time step')
            plt.ylabel('Prediction error')
            plt.show()

    def anomaly_likelihood(self, printer="n"):
        """
        Calculate the anomaly likelihood
        
        """ 
        L = np.zeros(self.T)

        # Model distribution of Prediction error values as "rolling normal distribution"
        W = 10 # Long-term window size
        w = 2  # Short-term window size
        epsilon = 10**-5 # Anomaly threshold

        for t in np.arange(self.T):
            W_range = np.arange(t,t-(W-1),-1)
            w_range = np.arange(t,t-(w-1),-1)

            # Long-term window mean
            mew_W = np.sum(self.S[W_range])/ W

            # Normal window standard deviation
            sigma_W = np.sum((self.S[W_range]-mew_W)**2) / (W-1)

            # Short-term window mean
            mew_w = np.sum(self.S[w_range])/ w 

            # Anomaly likelihood
            L[t] = 1 - Q_func((mew_w - mew_W)/sigma_W)

            # Threshold where likelihood very high [> 0.99999]
            Anomaly = L[L>=1-epsilon]   

        if printer == "Y":
            plt.plot(L)
            plt.xlabel('Time step')
            plt.ylabel('Anomaly likelihood')
            plt.show()


    def SP_phase1_initizlization(self,printer="n"):
        """
        Phase 1 - Compute initial potenital synapses based on a biased & random permanence distribution
        
        Zwischen variables
        -----------
        input_space : input data
        D           : dendtritic segments connecting input space and column space
        potentialSynapses: [column, column, mapping, mapping]

        Returns
        -----------
        potentialSynapses : intial mapping of column space with input space i.e. potential synapses
        
        """
        # Generate random permanence distribution for input region
        p_dist = np.random.normal(0,10, (self.R*2, self.R*2)) # about zero normal distribution
        rv = multivariate_normal([0, 0], [[self.R*2, 0], [0, self.R*2]]) # centred normal distribution
        
        x, y = np.mgrid[-self.R:self.R:1, -self.R:self.R:1]
        pos = np.empty(x.shape + (2,))
        pos[:, :, 0] = x; pos[:, :, 1] = y
        perm_dist = rv.pdf(pos)*p_dist+self.p_thresh # bias around p_thresh

        # create random initial mapping to input space with permanence values
        for idx, val in np.ndenumerate(self.columns): # loop through columns
            # reset input space to zero
            input_region = np.zeros((self.R*2,self.R*2))
            # set input region to perm_dist
            input_region = perm_dist
            # Specific number of initial input space synapses
            connect = np.random.choice([1, 0], size=((self.R*2,self.R*2)), p=[self.connectivity,1-self.connectivity])
            # computed potential connected synapses
            self.potentialSynapses[idx[0],idx[1],:,:] = connect*input_region
        
            # threshold with permanence value
            #if input_space[idx_con] < p_thresh:
            #    input_space[idx_con]==0
        
        if printer == "Y":
            plt.imshow(perm_dist)
            plt.title("Distribution of permanence values within initial input region")
            plt.colorbar()
            plt.show()

            plt.imshow(self.potentialSynapses[0,0,:,:])
            plt.title("Example of connection mapping of single column with initial input space")
            plt.colorbar()
            plt.show()    

    def SP_phase2_overlap(self,printer="n"):
        """
        Phase 2 - Compute the overlap with the current input for each column

        Described as the feedforward input to each mini-column

        TBC : boost factor 
        """ 

        # Generate random sparsely encoded input vector (later to be encoded with data)
        boost = 10 # boost factor
        input_vector = np.random.choice([1, 0], size=(self.N_input,self.N_input), p=[self.sparsity/100, (100-self.sparsity)/100])

        # Compute overlap (number of connected synapses with active inputs * boost factor)
        overlap = np.zeros((self.N_column,self.N_column))
        
        #pad input vector with zeros for indexing ease
        input_vector_pad = np.lib.pad(input_vector, self.R, 'constant', constant_values=(0,0))
        
        # loop through column indicies
        for idx, val in np.ndenumerate(self.potentialSynapses[:,:,0,0]):
            
            #isolate input region relative to columns potenital synapse region
            input_region=input_vector_pad[idx[0]:idx[0]+2*self.R:1,idx[1]:idx[1]+2*self.R:1]
            
            # compute overlaps 
            a = np.multiply(input_region,self.potentialSynapses[idx[0],idx[1],:,:]) 
            
            # anything matched is above zero *half-wave rectification
            a[a>0]=1 

            # sum connections
            overlap[idx[0],idx[1]] = overlap[idx[0],idx[1]] + np.sum(a) 
            
            # include boost
            #overlap[idx] = overlap[idx] * boost
        
        self.overlap = overlap
        
        if printer == "Y":
            plt.imshow(overlap)
            plt.colorbar()
            plt.title('Overlap')
            plt.show()


    def SP_phase3_competition(self,printer="n"):
        """
        Phase 3 - Compute the winning columns after inhibition step
    
        - Compute neighbourhoods for each column index
        - Calculate local activity minimum within this neighbourhood e.g. 10th smallest overlap score
        - Use this minimum local activity to threshold neighbourhood
        - Extract "winning" columns are saved
        
        Parameters
        ----------
        numActiveColumnsPerInhArea : number of active columns per inhibition area
        localAreaDensity           : desired density of active columns within a local inhibition area
        phi                        : inhibition radius
        
        """
        self.stimulusThreshold = 1
        numActiveColumnsPerInhArea = 3

        # localAreaDensity must be less than zero when numActiveColumnsPerInhArea is non-zero
        if numActiveColumnsPerInhArea != 0:
            localAreaDensity = -1
        else:
            localAreaDensity = 10 

        self.phi = 4 # inhibition radius

        # calculate active columns for each column neighbourhood using overlap (first calculate neighbours)
        activeColumns=[]
        # loop through columns
        for idx_current, val in np.ndenumerate(self.overlap):
            # indicies of inhibition neighbourhood for current column
            neighbourhood_overlap = []
            # loop through columns for each current column
            for idx_vs, val in np.ndenumerate(self.overlap):
                Euc_D = []
                # column is neighbour if non-identical and Euclidiean distance < phi i.e. within "inhibition radius"
                Euc_D = np.linalg.norm(np.array(idx_current)-np.array(idx_vs))
                if idx_vs != idx_current and Euc_D < self.phi:
                    neighbourhood_overlap.append(self.overlap[idx_vs])

            # indicies of sorted neighbourhood overlap
            sorted_neighbours_i = sorted(range(len(neighbourhood_overlap)), key=lambda k: neighbourhood_overlap[k])

            # minimum local activity 
            minLocalActivity = sorted_neighbours_i[-numActiveColumnsPerInhArea]

            # Apply global and local threshold - extract indicies of winning columns
            if self.overlap[idx_current] > self.stimulusThreshold and self.overlap[idx_current] >= minLocalActivity:
                activeColumns.append([idx_current[0],idx_current[1]])
        
        # Save active columns
        self.activeColumns=activeColumns
        
        # Load active column indicies into matrix for plotting
        activity = np.zeros_like(self.overlap)
        for i in activeColumns:
            activity[i[0],i[1]]=1
        
        self.activity = activity
        
        if printer == "Y":
            plt.imshow(activity)
            plt.colorbar()
            plt.title('Activity after inhibition')
            plt.show()


    def SP_phase4_update(self,printer="n"):
        """
        Phase 4 - Update synapse permanences and internal variables
    
        Updates permanences, boost and inihbition radii

        Parameters
        ----------
        
        
        TBC : 
            - moving averages aren't stored over time (perhaps this is needed)
            - definitions of later functions are ambiguous and might need double checking   
            - especially for recomputing the radius
            - boost function has nothing exponential to it
        """
        synPermActiveInc = 0.1   # active synpase permanence increment
        synPermInactiveDec = 0.1 # inactive synpase permanence increment
        s = 1.5
        
        for c in self.activeColumns: # loop through winning column indicies
            activeSynapses = self.potentialSynapses[c[0],c[1],:,:] > self.p_thresh # loop through active synapses
            for idx, val in np.ndenumerate(activeSynapses):
                if val == True: # if active
                    val = self.potentialSynapses[c[0],c[1],idx[0],idx[1]] + synPermActiveInc # increment permanence
                    self.potentialSynapses[c[0],c[1],idx[0],idx[1]] = np.minimum(1.0, val)
                else: # if inactive
                    val = self.potentialSynapses[c[0],c[1],idx[0],idx[1]] - synPermInactiveDec # decrement permanence
                    self.potentialSynapses[c[0],c[1],idx[0],idx[1]] = np.maximum(0.0, val)
        # Non-winning column synapse permanence values aren't modified
        
        activeDutyCycle=np.zeros_like(self.activity)
        boost=np.zeros_like(self.activity)
        minDutyCycle = 10 # initiate with high value
        
        activeDutyCycle = np.mean(self.activity)
        for idx_c, val_c in np.ndenumerate(self.activity): # loop through columns
            activeDutyCycleNeighbours = np.mean(neighbours(idx_c, self.activity, self.phi))
            
            #if a column is a frequent neighbourhood loser, its boost set to >1
            #if a column is a frequent neighbourhood winner, its boost set to <1
            boost[idx_c] = boostFunction(activeDutyCycle, activeDutyCycleNeighbours)
            
            overlapDutyCycle = np.mean(self.overlap[self.overlap > self.stimulusThreshold])
            minDutyCycle = np.minimum(activeDutyCycleNeighbours,minDutyCycle)
            
            # if a column's overlapDutyCycle is too low, its permanence values are boosted by scale factor s
            if overlapDutyCycle < minDutyCycle:
                self.potentialSynapses[idx_c[0],idx_c[1],:,:] += s
        
        # Inhibition radius recomputed
        inhibitionRadius = averageReceptiveFieldSize(self.potentialSynapses)
        
        if printer == 'Y':
            plt.imshow(self.activity)
            plt.colorbar()
            plt.show()


    def TM_phase1_context(self,t,printer='n',LEARNING_THRESHOLD=0.75):
        """
        Receive a set of active columns, evaluate them against predictions, and choose a set of active cells:
        
        a. For each active column, check for cells in the column that have an active distal dendrite segment 
        (i.e. cells that are in the “predictive state” from the previous time step), and activate
        them. 
        
        If no cells have active segments, activate all the cells in the column, marking this
        column as “bursting”. The resulting set of active cells is the representation of the input in the
        context of prior input.
        
        b. For each active column, learn on at least one distal segment. For every bursting column,
        choose a segment that had some active synapses at any permanence level. If there is no such
        segment, grow a new segment on the cell with the fewest segments, breaking ties randomly.
        On each of these learning segments, increase the permanence on every active synapse,
        decrease the permanence on every inactive synapse, and grow new synapses to cells that
        were previously active.
        
        TBC : What are the distal connections? Make sure of consistant labelling throughout code
        """
        
        self.matchingSegments = self.potentialSynapses[self.potentialSynapses>LEARNING_THRESHOLD]
        
        predictionsDistalActive = np.zeros_like(self.HTMlayers)
        predictionsDistalMatching = np.zeros_like(self.HTMlayers)
        # loop through active columns
        for idx_c, val_c in np.ndenumerate(self.HTMlayers[:,0]):
            if list(idx_c) in self.activeColumns: 
                # if any cell was predicted
                if np.count_nonzero(segmentsForColumn(idx_c, predictionsDistalActive[:,t-1])) > 0:
                    for idx_s in segmentsForColumn(column, activeSegments(t-1)):
                        
                        # Each of these cells is marked as a “winner”, making them presynaptic candidates for 
                        # synapse growth in the next time step 
                        activeCells(t).add(segment.cell[idx_s])
                        winnerCells(t).add(segment.cell[idx_s])

                        if LEARNING_ENABLED=="Y":
                            # For  each  of  these  correctly  active  segments
                            for synapse in segment.synapses: 
                                if synapse.presynapticCell in activeCells(t-1):
                                    #reinforce  the  synapses  that  activated the  segment
                                    synapse.permanence += PERMANENCE_INCREMENT
                                else:
                                    # punish   the   synapses   that   didn’t   contribute.
                                    synapse.permanence -= PERMANENCE_DECREMENT
                            
                            #If   the   segment   has   fewer   than SYNAPSE_SAMPLE_SIZE  active  synapses,
                            # grow new  synapses  to  a  subset  of  the 
                            # winner  cells  from  the previous time step to make up the difference
                            newSynapseCount = (SYNAPSE_SAMPLE_SIZE -numActivePotentialSynapses(t-1, segment))
                            growSynapses(segment, newSynapseCount)
                else:
                     burstColumn(idx_c, self.matchingSegments,t) # burst column
            else:
                 if np.count_nonzero(segmentsForColumn(idx_c, predictionsDistalMatching[:,t-1])) > 0:
                    punishPredictedColumn(idx_c, self.matchingSegments,t) # punish predicted column
        
        self.predictionsDistalActive = predictionsDistalActive
    
        
#     def TM_phase2_prediction(self,printer='n'):
#     """
#     Activate a set of dendrite segments: for every dendrite segment on every cell in the layer, count
#     how many connected synapses correspond to currently active cells (computed in step 1). If the
#     number exceeds a threshold, that dendrite segment is marked as active. The collection of cells
#     with active distal dendrite segments will be the predicted cells in the next time step.
#     """
#         if printer == 'Y':
#             plt.imshow(self.predictionsDistalActive)
#             plt.colorbar()
#             plt.show()
        


      

def burstColumn(column, matchingSegments,t):
    """
    If the column activation was unexpected, then each cell in the column becomes active 
        Select a winner cell and a learning segment for the column. 
        
    If any cells have a matching segment, select the best matching segment and its cell .
        Otherwise select the least used cell and grow a new segment on it 
        
    On the learning segment, reinforce the synapses that partially activated the segment, and punish the 
    synapses that didn’t contribute. 
    
    Then grow new synapses to a subset of the previous time step’s winner cells
    """
    for cell in column.cells:
        activeCells(t).add(cell)

    if segmentsForColumn(column, matchingSegments[t-1]).length > 0 :
        learningSegment = bestMatchingSegment(column)      
        winnerCell = learningSegment.cell   
    else:
        winnerCell = leastUsedCell(column)
        if LEARNING_ENABLED:
            learningSegment = growNewSegment(winnerCell)
            winnerCells(t).add(winnerCell)
    if LEARNING_ENABLED:
        for synapse in learningSegment.synapses:
            if synapse.presynapticCell in activeCells(t-1):
                synapse.permanence += PERMANENCE_INCREMENT
            else:
                synapse.permanence -= PERMANENCE_DECREMENT

        newSynapseCount = (SAMPLE_SIZE -numActivePotentialSynapses(t-1, learningSegment))
        growSynapses(learningSegment, newSynapseCount)
        

# ----- SUPPORT FUNCTIONS -----

def leastUsedCell(column):
    """
    The least used cell is the cell with the fewest segments. Find all the cells with the fewest segments 
    and choose one randomly
    """
    fewestSegments = INT_MAX
    for cell in column:
        fewestSegments = min(fewestSegments, cell.segments.length)

    leastUsedCells = []
    for cell in column:
        if cell.segments.length == fewestSegments:
            leastUsedCells.add(cell)

    return  chooseRandom(leastUsedCells)


def bestMatchingSegment(column):
    """
    The best matching segment is the “matching” segment with largest number of active potential synapses.
    """
    bestMatchingSegment = None
    bestScore = -1
    for segment in segmentsForColumn(column, matchingSegments(t-1)):
        if numActivePotentialSynapses(t-1, segment) > bestScore :
            bestMatchingSegment = segment
    bestScore = numActivePotentialSynapses(t-1, segment)
    return bestMatchingSegment

def growSynapses(segment, newSynapseCount):
    """
    When growing synapses, grow to a random subset of the previous time step’s winner cells. A segment 
    can only connect to a presynaptic cell once 
    """
    candidates = copy(winnerCells(t-1))
    while candidates.length > 0 and newSynapseCount > 0:
        presynapticCell = chooseRandom(candidates)
        candidates.remove(presynapticCell)
        alreadyConnected = false
        for synapse in segment.synapses:
            if synapse.presynapticCell == presynapticCell:
                alreadyConnected = true
        if alreadyConnected == false:
            newSynapse = createNewSynapse(segment, presynapticCell, INITIAL_PERMANENCE)
            newSynapseCount -= 1
            

def numActivePotentialSynapses(t,learningSegment):
    """
    
    """
    
    return val
    
    
def punishPredictedColumn(column, matchingSegments,t, LEARNING_ENABLED='Y'):
    """
    When a column with matching segments doesn’t become active, punish the synapses that 
    caused these segments to be “matching”
    """
    if LEARNING_ENABLED == 'Y':
        for segment in segmentsForColumn(column, matchingSegments[t-1]):
            for synapse in segment.synapses:
                if synapse.presynapticCell in activeCells(t-1):
                    synapse.permanence -= PREDICTED_DECREMENT
            
            
def segmentsForColumn(column,segments):
    """
    Given a set of segments, return the segments that are on a cell in this column.
    """
    segmentsInCell = []
    for idx_s, val in np.ndenumerate(segments):
        if idx_s == column:
            segmentsInCell.append(idx_s)
    
    return segmentsInCell
        
        
def neighbours(idx_c, columns,phi):
    """
    A list of all the columns that are within inhibitionRadius of column c. 
    
    returns neighbourhood values
    """
    neighbourhood = []
    for idx, val in np.ndenumerate(columns):
        Euc_D = np.linalg.norm(np.array(idx)-np.array(idx_c))
        if idx_c != idx and Euc_D < phi:
            neighbourhood.append(val)                                           
    return neighbourhood


def averageReceptiveFieldSize(synapses, connectedPerm=0.7):
    """
    Compute the radius of the average connected receptive field size of all the columns. 
    
    The connected receptive field size of a column includes only the connected synapses 
    (i.e. those with permanence values >= connectedPerm). 
    This is used to determine the extent of lateral inhibition between columns. 
    """
    max_R = 0
    columns_R=[]
    for idx_c, val_c in np.ndenumerate(synapses[:,:,0,0]): # loop through columns
        for idx_s, val_s in np.ndenumerate(synapses[idx_c[0],idx_c[1],:,:]): # loop through active synapses
            if val_s > connectedPerm:
                Euc_D = np.linalg.norm(np.array(idx_s)-np.array(idx_c))
            else:
                Euc_D = 0
            max_R = np.maximum(Euc_D,max_R) # maximum radius of connected synapses within a column
        columns_R.append(max_R) # store max radius for each column
        
    return np.mean(columns_R) #return average      
    
    
def boostFunction(activeDutyCycle, activeDutyCycleNeighbours,boostStrength = 2):
    """
    The boostFunction is an exponential function that depends on the difference between the active duty cycle
    of a column and the average active duty cycles of its neighbors. 
    --------------- vs -----------------
    The boost value is a positive scalar value. It is above one if the active duty cycle is above the mean 
    active duty cycles of neighboring columns. It is less than one if a column has higher active duty cycle 
    than its neighbors
    
    Returns
    -------
    boost : boost value of a column
    """
    boost=[]
    if activeDutyCycle > activeDutyCycleNeighbours :
        boost = np.random.random() + 1
    else:
        boost = np.random.random()

    return boost


def Q_func(x):
    """
    Normal distribution tail function approximation
    """
    return (1/2)*scipy.special.erf(x/np.sqrt(2))

        

# Run model and print outputs
babyHTM = babyHTM()
babyHTM.run_model()
```
